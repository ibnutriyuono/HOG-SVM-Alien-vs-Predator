```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def grayscale(img):
    gray = 0.2126 * img[:,:,0] + 0.7152 * img[:,:,1] + 0.2127 * img[:,:,2]
    gray_img = gray.astype(np.uint8)
    return gray_img
```

```python
orig = cv2.imread('./data/train/alien/8.jpg')
gray = grayscale(orig)
plt.imshow(gray, cmap='gray')
```
